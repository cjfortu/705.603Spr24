## Clean up
* Find and List number of blank entries and outliers/errors

**In the previous section we discussed the need to re-code date/time to preserve the ordering of year, and also the cyclical nature of days within a year as well as seconds within a day.**  
**We convert *trans_date_trans_time* to pandas datetime objects to allow easier separation of calendar day and time.**

```python
print(type(df['trans_date_trans_time'][0]))
print(df['trans_date_trans_time'][0])
```

**Now create a year column**

```python
df['year'] = df['trans_date_trans_time'].dt.year
print(df['year'])
```

**Now create a day of year column**  
**This is used for sine/cosine facetization, and will later be dropped.**

```python
df['day_of_year'] = df['trans_date_trans_time'].dt.day_of_year
print(df['day_of_year'])
```

**Now create a seconds in day column**  
**This is used for sine/cosine facetization, and will later be dropped.**

```python
df['sec_of_day'] = df['trans_date_trans_time'].dt.second +\
        60 * df['trans_date_trans_time'].dt.minute +\
        3600 * df['trans_date_trans_time'].dt.hour
print(df['sec_of_day'])
```

**Now we can expand both *day_of_year* and *sec_of_day* out to sine and cosine facets, in order to preserve their cyclical nature.**

```python
# First do day_of_year
df['doy_x'] =\
        np.cos(2 * np.pi * df.loc[df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        366)  # x component of day_of_year for leap years
df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'doy_x'] =\
        np.cos(2 * np.pi * df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        365)  # x component of day_of_year for non leap years
df['doy_y'] =\
        np.sin(2 * np.pi * df.loc[df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        366)  # y component of day_of_year for leap years
df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'doy_y'] =\
        np.sin(2 * np.pi * df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        365)  # y component of day_of_year for non leap years

# Now do sec_of_day
df['sod_x'] = np.cos(2 * np.pi * df['sec_of_day'] / 86400)  # x component of sec_of_day
df['sod_y'] = np.sin(2 * np.pi * df['sec_of_day'] / 86400)  # x component of sec_of_day

print(df['doy_x'])
print(df['doy_y'])
print(df['sod_x'])
print(df['sod_y'])
```

**Now we can drop *trans_date_trans_time* as well as the temporarily created columns.**

```python
df = df.drop(columns=['trans_date_trans_time', 'day_of_year', 'sec_of_day'])
```

**Now drop *cc_num*, *merchant*, *first*, *last*, and *street*, according to the discussion in the previous section.**

```python
df = df.drop(columns=['cc_num', 'merchant', 'first', 'last', 'street'])
```

***category* and *sex* are done, now move on to *zip*, *city*, *state*, and *job*.**

***zip* is done, now repeat the process for *city*, *state*, and *job*.**

```python
def ohe(df, params):
    for param in params:
        df['{}_cor'.format(param)] = 0
        valscounts = (df.loc[df['is_fraud'] == 1, '{}'.format(param)].value_counts() /\
               df['{}'.format(param)].value_counts()).sort_values(ascending=False) == 1.0
        truevals = []
        for zc, bval in valscounts.items():
            if bval == True:
                truevals.append(zc)

        df.loc[df['{}'.format(param)].isin(truevals), '{}_cor'.format(param)] = 1
        print(df['{}_cor'.format(param)].value_counts())
        df = pd.get_dummies(data=df, columns=['{}_cor'.format(param)])
        df = df.drop(columns=['{}'.format(param)])

    return df
        
df = ohe(df, ['zip', 'city', 'state', 'job'])
```
