## Clean up
* Find and List number of blank entries and outliers/errors

**In the previous section we discussed the need to re-code date/time to preserve the ordering of year, and also the cyclical nature of days within a year as well as seconds within a day.**  
**We convert *trans_date_trans_time* to pandas datetime objects to allow easier separation of calendar day and time.**

```python
print(type(df['trans_date_trans_time'][0]))
print(df['trans_date_trans_time'][0])
```

**Now create a year column**

```python
df['year'] = df['trans_date_trans_time'].dt.year
print(df['year'])
```

**Now create a day of year column**  
**This is used for sine/cosine facetization, and will later be dropped.**

```python
df['day_of_year'] = df['trans_date_trans_time'].dt.day_of_year
print(df['day_of_year'])
```

**Now create a seconds in day column**  
**This is used for sine/cosine facetization, and will later be dropped.**

```python
df['sec_of_day'] = df['trans_date_trans_time'].dt.second +\
        60 * df['trans_date_trans_time'].dt.minute +\
        3600 * df['trans_date_trans_time'].dt.hour
print(df['sec_of_day'])
```

**Now we can expand both *day_of_year* and *sec_of_day* out to sine and cosine facets, in order to preserve their cyclical nature.**

```python
# First do day_of_year
df['doy_x'] =\
        np.cos(2 * np.pi * df.loc[df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        366)  # x component of day_of_year for leap years
df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'doy_x'] =\
        np.cos(2 * np.pi * df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        365)  # x component of day_of_year for non leap years
df['doy_y'] =\
        np.sin(2 * np.pi * df.loc[df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        366)  # y component of day_of_year for leap years
df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'doy_y'] =\
        np.sin(2 * np.pi * df.loc[~df['trans_date_trans_time'].dt.is_leap_year, 'day_of_year'] /\
        365)  # y component of day_of_year for non leap years

# Now do sec_of_day
df['sod_x'] = np.cos(2 * np.pi * df['sec_of_day'] / 86400)  # x component of sec_of_day
df['sod_y'] = np.sin(2 * np.pi * df['sec_of_day'] / 86400)  # x component of sec_of_day

print(df['doy_x'])
print(df['doy_y'])
print(df['sod_x'])
print(df['sod_y'])
```

```python
def ohe(df, params):
    for param in params:
        df['{}_cor'.format(param)] = 0
        valscounts = (df.loc[df['is_fraud'] == 1, '{}'.format(param)].value_counts() /\
               df['{}'.format(param)].value_counts()).sort_values(ascending=False) == 1.0
        truevals = []
        for zc, bval in valscounts.items():
            if bval == True:
                truevals.append(zc)

        df.loc[df['{}'.format(param)].isin(truevals), '{}_cor'.format(param)] = 1
        print(df['{}_cor'.format(param)].value_counts())
        df = pd.get_dummies(data=df, columns=['{}_cor'.format(param)])
        df = df.drop(columns=['{}'.format(param)])

    return df
        
df = ohe(df, ['zip', 'city', 'state', 'job'])
```

## Split into 3 data sets for training, validation, and test (Explain your % for each)

**We will use %80 train, %10 validation, and %10 split.**  
**Although the dataset is very large, we will not use a smaller validation+test percentage because there are relatively few *is_fraud=1* data points.**  
**It is necessary to straify on *is_fraud*.**

```python
Y = df['is_fraud'].to_numpy()
X = df.drop(columns=['is_fraud']).to_numpy()
X_train, X_test, Y_train, Y_test =\
        train_test_split(X, Y, test_size=0.10, stratify=Y)
X_train, X_val, Y_train, Y_val =\
        train_test_split(X_train, Y_train, test_size=1/9, stratify=Y_train)

print('train/val/test data set sizes:\n')
print('X: {} {} {}'.format(X_train.shape, X_val.shape, X_test.shape))
print('Y: {} {} {}'.format(Y_train.shape, Y_val.shape, Y_test.shape))
```

```python
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

# we will use a random forest
for min_samples_split in range(2, 11):
    n_estimators = 100
    max_features = 'sqrt'
    model = RandomForestClassifier(n_jobs = -1,
                                      n_estimators = n_estimators,
                                      min_samples_split = min_samples_split,
                                      max_features = max_features,
                                      verbose = 1)
    model.fit(X_train, Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_val = model.predict(X_val)

    # we will use F1 score as the first performance metric
    F1_train = f1_score(Y_train, Y_pred_train)
    F1_val = f1_score(Y_val, Y_pred_val)

    # ROC AUC score will be the second performance metric
    ROCAUC_train = roc_auc_score(Y_train, Y_pred_train)
    ROCAUC_val = roc_auc_score(Y_val, Y_pred_val)

    print('min_samples_split: {}'.format(min_samples_split))
    print('F1_score train: {}, val: {}\nROCAUC train: {}, val: {}'.\
          format(F1_train, F1_val, ROCAUC_train, ROCAUC_val))
```

**regularizing by increasing *min_samples_split* lowered the training performance but did not raise the testing performance.**  
  
**Nonetheless we will use *min_samples_split*=10 for testing, in order to keep training and testing performance closer together.**

```python
n_estimators = 100
min_samples_split = 10
max_features = 'sqrt'
model = RandomForestClassifier(n_jobs = -1,
                                  n_estimators = n_estimators,
                                  min_samples_split = min_samples_split,
                                  max_features = max_features,
                                  verbose = 1)
model.fit(X_train, Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

# we will use F1 score as the first performance metric
F1_train = f1_score(Y_train, Y_pred_train)
F1_test = f1_score(Y_test, Y_pred_test)

# ROC AUC score will be the second performance metric
ROCAUC_train = roc_auc_score(Y_train, Y_pred_train)
ROCAUC_test = roc_auc_score(Y_test, Y_pred_test)

print('min_samples_split: {}'.format(min_samples_split))
print('F1_score train: {}, test: {}\nROCAUC train: {}, test: {}'.\
      format(F1_train, F1_test, ROCAUC_train, ROCAUC_test))
```

**Both F1 and ROC AUC scores are on the same scale, and are assessed similarly.**  

**>0.8 is often considered very good, while >0.9 is often considred excellent.**  

**The model training performance is >0.9 while in testing it is >0.8.  I'll consider this effective in capturing the relationship between transaction data and fraud.**
