# Purpose and Justification

It seems current object detection/tracking models require the object to be ascertainable in a single frame (ex: in a single video frame you can tell if there is a person in the video or not.  Over the course of the video, you can follow the person https://arxiv.org/abs/1711.11248v3).  There are cases where the object is not ascertainable in a single frame, but its motion is trackable over the course of video (imagine a video of many identical marbles, but only one is moving differently than the others).  


Computer vision capability can be more comprehensive with the ability to detect indistinguishable movers; this has military, law enforcement, and medical applications.


# Domain Expertise

This is context-dependent.  Applications will have human analysts trained to understand the phenomenology of different domains.  The common capability is that of humans being able to notice objects moving, which are otherwise indistinguishable from each other.

# Overall Design

The model requires a 3D CNN for video classification, and a 3D extention to YOLO.  Training will undoubtedly require GPU(s), while deployment could rely on multiple CPUs (give each spatial slice of a video to a CPU process).
