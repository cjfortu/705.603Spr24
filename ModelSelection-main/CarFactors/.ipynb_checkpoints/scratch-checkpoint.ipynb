```python
df = df.dropna()
```

```python
df['manumod'] = df.loc[:, ['manufacturer_name', 'model_name']].apply(lambda row:\
                '_'.join(row.values.astype(str)), axis=1)
```

```python
df
```

```python
pd.set_option('display.max_rows', None)
# df['manumod'].value_counts()
display(df['manumod'].value_counts().size)
pd.set_option('display.max_rows', 10)
```

```python
for item in df['manumod'].value_counts().iteritems():
    print(item)
```

```python
manumod_switch = {}
manumodnum = 0
for manumod in df['manumod'].value_counts().items():
    if manumod[1] < 10:
        manumod_switch[manumod[0]] = df['manumod'].value_counts().size
#         if 'other' in manumod_switch.keys():
#             pass
#         else:
#             manumod_switch['other'] = df['manumod'].value_counts().size
    elif manumod[0] in manumod_switch.keys():
        pass
    else:
        manumod_switch[manumod[0]] = manumodnum
        manumodnum += 1

df['manumod_num'] = df['manumod'].map(manumod_switch)
```

```python
df = df[['manufacturer_num', 'manufacturer_name', 'transmission', 'color', 'odometer_value', 'year_produced', 'engine_type',\
         'engine_capacity', 'body_type', 'has_warranty', 'drivetrain', 'price_usd',\
         'duration_listed']]
```

```python
body_switch = {
    'universal': 0,
    'hatchback': 1,
    'cabriolet': 2,
    'coupe': 3,
    'sedan': 4,
    'liftback': 5,
    'suv': 6,
    'minivan': 7,
    'van': 8,
    'pickup': 9,
    'minibus': 10,
    'limousine': 11
}
```

```python
df = df.replace({'body_type': body_switch})
```

```python
df
```

```python
normcols = ['odometer_value', 'year_produced', 'engine_capacity', 'price_usd']
for col in normcols:
    df[col] = df[col] / df[col].max()
```
